```python
import torch

class AdamOptimizer:
    def __init__(self, params, lr=1e-3,
                 beta1=0.9, beta2=0.999, eps=1e-8):
        self.params =params
        self.lr = lr
        self.beta1 = beta1
        self.beta2 = beta2
        self.eps = eps
        self.m = [torch.zeros_like(p) for p in params]
        self.v = [torch.zeros_like(p) for p in params]
        self.len_params = len(params)
        self.t = 0

    def step(self, grads):
      self.t+=1
      for i in range(self.len_params):
        self.m[i] = self.beta1*self.m[i]+(1-self.beta1)*grads[i]
        self.v[i] = self.beta2*self.v[i]+(1-self.beta2)*grads[i]**2
        if self.t==1:
          self.m[i] = self.m[i]/(1-self.beta1)
          self.v[i] = self.v[i]/(1-self.beta2)
        self.params[i].data = self.params[i].data-self.lr*self.m[i]/(torch.sqrt(self.v[i])+self.eps)

    def zero_grad(self):
      for i in range(self.len_params):
        self.params[i].grad = None
```
