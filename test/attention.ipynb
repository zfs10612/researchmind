```python
import torch
import math
def scaled_dot_product_attention(Q, K, V, mask=None):
  d_k = Q.shape[-1]
  dot_product = torch.matmul(Q, K.transpose(-2,-1))/math.sqrt(d_k)
  if mask is not None:
    dot_product = torch.masked_fill(dot_product, mask, -1e9)
  weight = torch.softmax(dot_product, dim = -1)
  return torch.matmul(weight, V)
```

```python
import torch

B, H, S, D = 2, 4, 8, 16
Q = torch.randn(B, H, S, D)
K = torch.randn(B, H, S, D)
V = torch.randn(B, H, S, D)

# Test 1: no mask
out = scaled_dot_product_attention(Q, K, V)
assert out.shape == (B, H, S, D), f"Got {out.shape}"

# Test 2: causal mask
mask = torch.triu(torch.ones(S, S), diagonal=1).bool()
out_masked = scaled_dot_product_attention(Q, K, V, mask=mask)
assert out_masked.shape == (B, H, S, D), f"Got {out_masked.shape}"

print("All assertions passed ✅")
```
